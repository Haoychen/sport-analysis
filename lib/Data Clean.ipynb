```python
import xml.dom.minidom
```

```python
# read xml file
DomTree = xml.dom.minidom.parse("f24-98-2012-429860-eventdetails.xml")
```

```python
# get the root
DomData = DomTree.documentElement
```

```python
timestamp = DomData.getAttribute('timestamp')
```

```python
Game = DomData.getElementsByTagName('Game')
```

```python
Game_id = Game[0].getAttribute('id')
away_team_id = Game[0].getAttribute('away_team_id')
away_team_name = Game[0].getAttribute('away_team_name')
competition_id = Game[0].getAttribute('competition_id')
competition_name = Game[0].getAttribute('competition_name')
game_date = Game[0].getAttribute('game_date')
home_team_id = Game[0].getAttribute('home_team_id')
home_team_name = Game[0].getAttribute('home_team_name')
matchday = Game[0].getAttribute('matchday')
period_1_start = Game[0].getAttribute('period_1_start')
period_2_start = Game[0].getAttribute('period_2_start')
season_id = Game[0].getAttribute('season_id')
seanson_name = Game[0].getAttribute('season_name')
```

```python
Events = Game[0].getElementsByTagName('Event')
```

```python
ids = []
event_id = []
type_id = []
period_id = []
mins = []
secs = []
player_id = []
team_id = []
outcome = []
x_s = []
y_s = []
timestamps = []
last_modified = []
for event in Events:
    ids.append(event.getAttribute('id'))
    event_id.append(event.getAttribute('event_id'))
    type_id.append(event.getAttribute('type_id'))
    period_id.append(event.getAttribute('period_id'))
    mins.append(event.getAttribute('min'))
    secs.append(event.getAttribute('sec'))
    player_id.append(event.getAttribute('player_id'))
    team_id.append(event.getAttribute('team_id'))
    outcome.append(event.getAttribute('outcome'))
    x_s.append(event.getAttribute('x'))
    y_s.append(event.getAttribute('y'))
    timestamps.append(event.getAttribute('timestamp'))
    last_modified.append(event.getAttribute('last_modified'))
#    Qs = event.getElementsByTagName('Q')
#    for Q in Qs:
#        print Q.getAttribute('id')
```
